```python
import re
import requests
from bs4 import BeautifulSoup as bs
res = requests.get('http://www.startravel.com.tw/outbound/JP/index.html')
res.encoding = 'UTF-8'

soup = bs(res.text)
title = []
for ele in soup.select('.contentText a'):
    if len(ele.text) > 9 :
        title.append(ele.text)
projectList = []
fullDic = {}
for index in range(0,len(title)):
    fullDic['title'] = title[index]
    tempFullDic = fullDic.copy()
    projectList.append(tempFullDic)
    
for ele in projectList:
    print ele['title']
```
